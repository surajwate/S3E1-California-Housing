```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../input/train_folds.csv")
df.head()
```

```python
df.size
```

```python
fold = 0
df_train = df[df.kfold != fold].reset_index(drop=True)
df_valid = df[df.kfold == fold].reset_index(drop=True)
df_train.drop(columns=["id", "kfold"], axis=1, inplace=True)
df_valid.drop(columns=["id", "kfold"], axis=1, inplace=True)
```

```python
# Split the target and features
X_train = df_train.drop("MedHouseVal", axis=1)
y_train = df_train["MedHouseVal"]
X_valid = df_valid.drop("MedHouseVal", axis=1)
y_valid = df_valid["MedHouseVal"]
```

```python
features = X_train.columns
```

```python
# Standardize the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_valid = scaler.transform(X_valid)
```

```python
# Build the linear regression model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_valid)
```

```python
# Check the performance of the model
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_valid, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_valid, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R^2 Score: {r2}")
```
