## Binning the target variable

### 1. Equal-Width Binning

- **Description**: The range of the target variable is divided into equal-width intervals.
- **Use Case**: Suitable when you want to maintain the original range structure in the bins.

```python
num_bins = 5
# df['MedHouseVal_binned'] = pd.cut(df['MedHouseVal'], num_bins, labels=range(num_bins))
df['MedHouseVal_binned'] = pd.cut(df['MedHouseVal'], num_bins, labels=False)

# Get the value counts and sort by the bin labels
df['MedHouseVal_binned'].value_counts().sort_index().plot(kind='bar')
plt.show()
df['MedHouseVal_binned'].value_counts().sort_index()
```

### 2. Equal-Frequency Binning (Quantile Binning)

- **Description**: Each bin will contain approximately the same number of samples.
- **Use Case**: Useful when you want each bin to have the same number of samples, which can help in stratified sampling.

```python
num_bins = 5
df['MedHouseVal_binned'] = pd.qcut(df['MedHouseVal'], q=num_bins, labels=range(num_bins))
# Get the value counts and sort by the bin labels
df['MedHouseVal_binned'].value_counts().sort_index().plot(kind='bar')
plt.show()
df['MedHouseVal_binned'].value_counts().sort_index()
```

### 3. Custom Binning

- **Description**: Define the specific bin edges manually, based on domain knowledge or data distribution.
- **Use Case**: When you have a knowledge of the data distribution or want to create bins with specific ranges.

```python
# Define bin edges
bins = [0, 1, 2, 3, 4.841, 5.1]
# Bin labels
labels = [0, 1, 2, 3, 4]

df['MedHouseVal_binned'] = pd.cut(df['MedHouseVal'], bins=bins, labels=labels, include_lowest=True)
# Get the value counts and sort by the bin labels
df['MedHouseVal_binned'].value_counts().sort_index().plot(kind='bar')
plt.show()
df['MedHouseVal_binned'].value_counts().sort_index()
```

### 4. K-Means Binning

- **Description**: Use K-Means clustering to create bins based on centroids.
- **Use Case**: Useful when you want to create bins based on the similarity of target values.

```python
from sklearn.cluster import KMeans

num_bins = 5

kmeans = KMeans(n_clusters=num_bins, random_state=42)
df['MedHouseVal_binned'] = kmeans.fit_predict(df[['MedHouseVal']])

# Get the value counts and sort by the bin labels
df['MedHouseVal_binned'].value_counts().sort_index().plot(kind='bar')
plt.show()
df['MedHouseVal_binned'].value_counts().sort_index()
```
